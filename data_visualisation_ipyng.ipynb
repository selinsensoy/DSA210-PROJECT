```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# 1. Daily Average Wind Speed by City
plt.figure(figsize=(12,6))
sns.lineplot(data=daily_wind, x='date', y='wind_speed', hue='city')
plt.title('Daily Average Wind Speed by City')
plt.xlabel('Date')
plt.ylabel('Wind Speed')
plt.legend(title='City')
plt.show()

# 2. Daily Total Fuel Consumption by City
plt.figure(figsize=(12,6))
sns.lineplot(data=daily_fuel, x='date', y='fuel_consumption_kg', hue='city')
plt.title('Daily Total Fuel Consumption by City')
plt.xlabel('Date')
plt.ylabel('Fuel Consumption (kg)')
plt.legend(title='City')
plt.show()

# 3. Scatter plot with regression line: Fuel Consumption vs Wind Speed by city
plt.figure(figsize=(10,6))
sns.lmplot(data=merged_wind_data, x='wind_speed', y='fuel_consumption_kg', hue='city', aspect=1.5, scatter_kws={'alpha':0.5})
plt.title('Fuel Consumption vs Wind Speed (with regression)')
plt.xlabel('Wind Speed')
plt.ylabel('Fuel Consumption (kg)')
plt.show()

# 4. Jointplot (scatter + histograms + regression) for overall data
sns.jointplot(data=merged_wind_data, x='wind_speed', y='fuel_consumption_kg', kind='reg', height=8, scatter_kws={'alpha':0.4})
plt.suptitle('Joint Plot: Fuel Consumption vs Wind Speed')
plt.tight_layout()
plt.show()



```
